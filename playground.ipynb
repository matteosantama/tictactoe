```python
import pandas as pd

from src import TicTacToe, RLPlayer, RandomPlayer
```

```python
tictactoe = TicTacToe(verbose=False)

rl_player = RLPlayer('rl')
random_player = RandomPlayer('random')

# tictactoe.register(rl_player)
tictactoe.register(random_player)
tictactoe.register(rl_player)
```

```python
num_games = 30000
wins = []
losses = []
ties = []

for episode in range(num_games):
    winner = tictactoe.play()
    wins.append(winner == rl_player)
    losses.append(winner == random_player)
    ties.append(winner is None)
    tictactoe.reset()
```

```python
wins = pd.Series(wins, name='wins')
losses = pd.Series(losses, name='losses')
ties = pd.Series(ties, name='ties')

window = 500
wins.rolling(window).mean().plot(legend=True)
losses.rolling(window).mean().plot(legend=True)
ties.rolling(window).mean().plot(legend=True)
```
