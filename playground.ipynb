```python
import pandas as pd

from src import TicTacToe, RLPlayer, RandomPlayer
```

```python
rl = RLPlayer('rl')
rando = RandomPlayer('random')
adversary = RLPlayer('adversary')

num_games = 30000
window = 1000
```

```python
game1 = TicTacToe(verbose=False)
game1.register(rl)
game1.register(rando)

p1 = []
p2 = []
ties = []

for episode in range(num_games):
    winner = game1.play()
    p1.append(winner == game1.players[0])
    p2.append(winner == game1.players[1])
    ties.append(winner is None)
    game1.reset()
    
p1 = pd.Series(p1, name=game1.players[0])
p2 = pd.Series(p2, name=game1.players[1])
ties = pd.Series(ties, name='tie %')

p1.rolling(window).mean().plot(legend=True)
p2.rolling(window).mean().plot(legend=True)
ties.rolling(window).mean().plot(legend=True)
```
